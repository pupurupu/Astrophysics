```python
def lk(x, x_, k, s):
    l =1
    for j in range(s-n, s+1):
        if j!=k:
            l *= (x_-x[j])/(x[k]-x[j])
    return(l)

def Ln(L_, n, x_, x, y):
    L = 0

    for i in range(len(x)-1):
        if x[i] == x_:
            L = y[i]
            L_.append(L)
            break

        elif x[i+1] == x_:
            L = y[i+1]
            L_.append(L)
            break

        elif (x[i] < x_) and (x_ < x[i + 1]):
            s = i+n_
            for k in range(s-n, s+1):
                L += y[k] * lk(x, x_, k, s)
            L_.append(L)
```
